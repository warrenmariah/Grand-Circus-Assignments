## Question 4: What is the range of years when the houses are sold?

```python
import pandas as pd
df = pd.read_csv('/home/mariahwarren/AmesHousing.csv/AmesHousing.csv')
year_sold = df['Yr Sold'].max() - df['Yr Sold'].min()
print("Range of years when houses are sold:", year_sold)
```
