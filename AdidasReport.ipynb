```python
def is_outlier(column: pd.Series):
    Q1=column.quantile(0.25)
    Q3=column.quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    return ~column.between(lower_bound,upper_bound)
```

```python
df[df.select_dtypes(include=np.number).apply(is_outlier)==False]
```

```python
plt.figure(figsize=(16,10))
sns.displot(data=df,x='Price per Unit',y='Total Sales',col='Product')
plt.suptitle('Best Selling Products by Price per Unit',fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(12,8))
sns.barplot(data=df,x='Retailer',y='Total Sales',hue='Region')
plt.title('Top Selling Retailers By Region', fontsize=16)
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(10,6))
sns.barplot(data=df,x='Sales Method',y='Total Sales')
plt.title('Total Sales by Sale Method',fontsize=16)
plt.tight_layout()
plt.show()
```
